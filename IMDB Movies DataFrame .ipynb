### Re-classify the movies 

```python
data.columns
```

```python
def rating(rating):
    if rating >=7.0 :
        return "Excellent"
    elif rating >=6.0 :
        return "Good"
    else : 
        return "Average"
    
        
```

```python
data['rating_per_categry'] = data['Rating'].apply(rating)
data.head()
```

```python
data.groupby('rating_per_categry')['Rating'].count()
```

### Unique values from Genre 

```python
data.columns
```

```python
data['Genre']
```

```python
list1=[]
for text in data['Genre']:
    list1.append(text.split(','))
list1
```

```python
list2 = []
for item in list1:
    for item1 in item:
        list2.append(item1)
list2
```

```python
unique = []
for item in list2:
    if item not in unique:
        unique.append(item)
unique
        
        
```
